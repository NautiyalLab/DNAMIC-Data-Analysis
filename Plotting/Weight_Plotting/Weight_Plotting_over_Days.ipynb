### Weight Plotting

```python

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python

weight_df = pd.read_csv("DNAMIC_Group2_AvA.csv")
weight_df = weight_df.drop(['Subject #'], axis=1)
```

```python
weight_df.head()
```

```python
weight_df = weight_df.set_index('Box Number')

plot_df = weight_df.T
```

```python
plot_df  # columns is int NOT string
# plot_df.columns
```

```python
# # Averaging over first 5 / next 5 to get the mean and sem

plot_df["Adult Avg"] = plot_df.iloc[:, :5].mean(axis=1)  # average over first 5 
plot_df['Adol Avg'] = plot_df.iloc[:, 5:10].mean(axis=1) # average over next 5
plot_df['Adult Sem'] = plot_df.iloc[:, :5].sem(axis=1, ddof=1)
plot_df['Adol Sem'] = plot_df.iloc[:, 5:10].sem(axis=1, ddof=1)
plot_df
```

### Weight Plotting

```python
number = plot_df.columns

# fig, ax = plt.subplots(figsize=(8,5))
fig, ax = plt.subplots(figsize=(12,9))
# range(1,5)
for i in range(1,6):
    box_number = i   # 
    plt.plot(plot_df[box_number], c='red', alpha = 0.15)

for j in range(6,11):
    box_number = j
    plt.plot(plot_df[box_number], c='blue', alpha = 0.15)

plt.errorbar(x=plot_df.index, y=plot_df["Adult Avg"], yerr=plot_df['Adult Sem'], c='red')
plt.errorbar(x=plot_df.index, y=plot_df["Adol Avg"], yerr=plot_df['Adol Sem'], c='blue')
    
plt.legend(['755', '756', '757', '758', '762', '792', '793', '794', '797', '798', "Adult Avg", "Adol Avg"], loc='upper center', bbox_to_anchor=(1.12, 1.02))
plt.title("'Group 2' Weights Over Days")
plt.xlabel("Days")
plt.ylabel("Weights (g)")

plt.xticks(rotation=90)

# paradigm_position = ['2019-09-18','2019-09-19','2019-09-22','2019-10-6','2019-10-9'] 
# t_l = []
# for i in paradigm_position:
#     dt = pd.to_datetime(i)
#     t_l.append(dt)

# # # plot dates of paradigms (vertical lines)
# for p in t_l:
#     ax.axvline(x=p, color='k', linestyle='--', dashes=(3, 5), linewidth=2, alpha=0.5)

# # xposition = [pd.to_datetime('2010-01-01'), pd.to_datetime('2015-12-31')]
# # for xc in xposition:
# #     ax.axvline(x=xc, color='k', linestyle='-')

    


plt.tight_layout()
plt.savefig("Group 2 - Weights Over Days (until 10-28)")

# plt.show()

```
